```python
import requests
from bs4 import BeautifulSoup
import uuid
import pandas as pd
import re
```

```python
url = 'https://id.wikipedia.org/wiki/Daftar_gubernur_dan_wakil_gubernur_di_Indonesia'

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')
```

```python
table = soup.find('table', {'class': 'wikitable'})
rows = table.find_all('tr')[1:]
```

```python
for row in rows:
      col = row.find_all('td')
      print([c.get_text(strip=True) for c in col])
```

# Answer Question 1

```python
data_gov = []
link_wiki = []
for row in rows:
      col = row.find_all('td')
      if len(col) >= 2:
            gov_id = str(uuid.uuid4())
            gov_name = col[3].get_text(strip=True)
            province = col[0].get_text(strip=True)
            prov_split = province.split('Daftar')[0]
            link_gov = col[3].find('a')
            wikipedia_link = f"https://id.wikipedia.org{link_gov['href']}" if link_gov else' N/A'
            link_wiki.append(wikipedia_link)
            gov_id
            data_gov.append([gov_id, gov_name, prov_split, wikipedia_link])
df_gov= pd.DataFrame(data_gov, columns=['ID', 'Governor Name', 'Province', 'Personal Wikipedia Link'])
df_gov.to_csv('answer_question_1.csv', index=False)
df_gov
```

# Answer Question 2

```python
gov_details = []
for link in link_wiki:
      if link == 'N/A':
        continue
      response = requests.get(link)
      soup_link = BeautifulSoup(response.text, 'html.parser')

      name_gov = soup_link.find('span', {'class': 'mw-page-title-main'})
      name_governor = name_gov.get_text(strip=True) if name_gov else 'N/A'


      table_link = soup_link.find('table', {'class': 'infobox vcard'})

      if table_link:
            birth_th = table_link.find('th', string='Lahir')
            if birth_th:
                  birth_td = birth_th.find_next_sibling('td')
                  if birth_td:
                        # Date of Birth
                        date = birth_td.get_text(strip=True).split('(')[0].strip()
                        group = re.search(r'\d{1,2} \w+ \d{2,4}', date)
                        birth_date = group.group() if group else 'N/A'

                        #Place of Birth
                        place = [a.get_text(strip=True) for a  in birth_td.find_all('a')]
                        places = ', '.join(place)
                        place_birth = re.sub(r'^\[\d+\],?s*|,?\s*\[\d+\]$', '', places) if places else 'N/A'
                  else:
                        birth_date, place_birth = 'N/A', 'N/A'
            else:
                  birth_date, place_birth = 'N/A', 'N/A'
      else:
            birth_date, place_birth = 'N/A', 'N/A'


      gov_details.append([name_governor, place_birth, birth_date, link])
df_gov_details = pd.DataFrame(gov_details, columns=['Governor Name', 'Place of Birth', 'Date of Birth', 'Wikipedia Personal Link'])
df_gov_details_2 = pd.concat([df_gov['ID'], df_gov_details], axis=1)
df_gov_details_2.to_csv('answer_question_2.csv', index= False)
df_gov_details_2
```
